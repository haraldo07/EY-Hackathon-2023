```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def imshow(title = "Image", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
image=cv2.imread('Aamai.jpg')
imshow("Aamai",image)
```

```python
plt.imshow(image)
```

```python
image1=cv2.resize(image,None,fx=0.5,fy=0.5)
imshow("Resizing the image",image1)
```

```python
image2=cv2.resize(image,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)
imshow("Resizing the image",image2)
```

```python
image3=cv2.resize(image,None,fx=2,fy=2,interpolation=cv2.INTER_NEAREST)
imshow("Resizing the image",image3)
```

```python
image3=cv2.resize(image,None,fx=5,fy=5,interpolation=cv2.INTER_LANCZOS4)
imshow("Resizing the image",image3)
```

```python
image4=cv2.resize(image,None,fx=7,fy=7,interpolation=cv2.INTER_CUBIC)
imshow("Resizing the image",image4)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import colors,cm
```

```python
image_cvt= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
r,g,b=cv2.split(image_cvt)
fig=plt.figure()
axis=fig.add_subplot(1,1,1,projection='3d')
pixel_colors = image_cvt.reshape((np.shape(image_cvt)[0]*np.shape(image_cvt)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
axis.scatter(r.flatten(),g.flatten(),b.flatten(),facecolors=pixel_colors,marker='*')
axis.set_xlabel('Red')
axis.set_ylabel('Green')
axis.set_zlabel('Blue')
```

```python
image = cv2.imread('Aamai.jpg')
flattened_image = image.reshape((-1, 3))
unique_colors = set(map(tuple, flattened_image))
unique_colors = np.array(list(unique_colors))
imshow('Original Image', image)
for i, color in enumerate(unique_colors):
    color_image = np.full((100, 100, 3), color, dtype=np.uint8)
    imshow(f'Color {i}', color_image)
```
