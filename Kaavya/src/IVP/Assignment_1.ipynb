```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('RollCall.xlsx')
```

```python
df.info()
```

```python
import random
weight=[]
height=[]
for i in range(72):
    h=round(random.uniform(5.4,6.1),2)
    height.append(h)
    w=random.randint(50,75)
    weight.append(w)
    
    
```

```python
df['Weight']=weight
df['Height']=height
```

```python
import seaborn as sns
sns.kdeplot(df['Height'])
```

```python
sns.kdeplot(df['Weight'])
```

```python
print("Statistical Inference of Height")
print("Mean:",df['Height'].mean())
print("Median:",df['Height'].median())
print("Mode:",df['Height'].mode()[0])
print("Standard Deviation:",df['Height'].std())
print("Variance:",df['Height'].var())
print("Skew:",df['Height'].skew())
print("Kurtosis:",df['Height'].kurtosis())
```

```python
print("Statistical Inference of Weight")
print("Mean:",df['Weight'].mean())
print("Median:",df['Weight'].median())
print("Mode:",df['Weight'].mode()[0])
print("Standard Deviation:",df['Weight'].std())
print("Variance:",df['Weight'].var())
print("Skew:",df['Weight'].skew())
print("Kurtosis:",df['Weight'].kurtosis())
```
